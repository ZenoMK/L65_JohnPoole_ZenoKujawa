```python
with plt.style.context(spstyle.get_style('nature-reviews')):
    fig, ax = plt.subplots(ncols =1, sharey = True)
    
model_means = [summary_dec["Argmax_Model_Accuracy"][1],summary_dec["altUpwards_Model_Accuracy"][1],summary_dec["Upwards_Model_Accuracy"][1],summary_dec["Random_Model_Accuracy"][1]]
true_means = [summary_dec["Argmax_True_Accuracy"][1],summary_dec["altUpwards_True_Accuracy"][1],summary_dec["Upwards_True_Accuracy"][1],summary_dec["Random_True_Accuracy"][1]]

model_std = [summary_dec["Argmax_Model_Accuracy"][2],summary_dec["altUpwards_Model_Accuracy"][2],summary_dec["Upwards_Model_Accuracy"][2],summary_dec["Random_Model_Accuracy"][2]]
true_std = [summary_dec["Argmax_True_Accuracy"][2],summary_dec["altUpwards_True_Accuracy"][2],summary_dec["Upwards_True_Accuracy"][2],summary_dec["Random_True_Accuracy"][2]]


plt.bar(np.arange(len(model_means))-0.2,true_means, 0.4,label = "Ground truth")
plt.bar(np.arange(len(model_means))+0.2,model_means, 0.4, label = "Model")
plt.xticks(np.arange(len(model_means)),["Argmax", "altUpwards","Upwards","Random"])
[print("&$"+str(model_means[i])+"\pm"+str(model_std[i])+"$") for i in range(len(model_means))]
print("TRUE BELOW")
[print("&$"+str(true_means[i])+"\pm"+str(true_std[i])+"$") for i in range(len(model_means))]

# taken from here: https://www.geeksforgeeks.org/add-error-bars-to-a-matplotlib-bar-plot/
# https://stackoverflow.com/questions/40020274/pyplot-errorbar-keeps-connecting-my-points-with-lines
plt.errorbar(np.arange(len(model_means))-0.2,true_means,yerr = true_std,color = "black",capsize = 5, ls = "None")
plt.errorbar(np.arange(len(model_means))+0.2,model_means,yerr = model_std,color = "black",capsize = 5, ls = "None")
plt.legend()
plt.ylabel("Fraction of valid trees")
plt.title("Accuracies for different sampling methods for |V(G)| = 5 (DFS)")
plt.savefig("DFS_n5.png",dpi = 1200)
plt.show()
```
